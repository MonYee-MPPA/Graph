```python
# link to GitHub
linkGoogle="https://github.com/DACSS-CSSmeths/Networks_intro/raw/refs/heads/main/casesData/SeattleTopData.xlsx"
```

```python
import pandas as pd
edge_attr=pd.read_excel(linkGoogle,sheet_name='attributes')
edge_attr
```

```python
#Draw directed graph(asymmetric relationship)

edgelist_d= pd.read_excel(linkGoogle,
                           sheet_name='edges') # name of the sheet

# see the data frame
edgelist_d
```

```python
import networkx as nx
graph_edgelist_d=nx.from_pandas_edgelist(edgelist_d,edge_attr=True,
                                         create_using=nx.DiGraph) # here!!
# see edges
graph_edgelist_d.edges.data()
```

```python
#Building dictionary for attributes
{n:s for n,s in zip(edge_attr.name,edge_attr.male)}
{n:s for n,s in zip(edge_attr.name,edge_attr.followers)}
```

```python
attrDic_male={n:s for n,s in zip(edge_attr.name,edge_attr.male)}
attrDic_followers={n:s for n,s in zip(edge_attr.name,edge_attr.followers)}
nx.set_node_attributes(graph_edgelist_d,attrDic_male , "male")
nx.set_node_attributes(graph_edgelist_d,attrDic_followers , "followers")
```

```python
graph_edgelist_d.nodes.data()
G2=graph_edgelist_d
G2
```

```python
#Find missing folder path. 
import os

examplesFolder = "Graph/"
os.makedirs(examplesFolder, exist_ok=True)  # Create the folder if it doesn't exist

nx.write_graphml(G2, examplesFolder + "css_G2.graphml", named_key_ids=True)
```
