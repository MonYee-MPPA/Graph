```python
# reading in
#We can download the adjacency from my Excel file from GitHub:
import pandas as pd
LinkToData="https://github.com/DACSS-CSSmeths/Networks_intro/raw/refs/heads/main/casesData/dataFigueroa.xlsx"
FigueData = pd.read_excel(LinkToData,sheet_name='Relationships',
                          index_col=0) #notice!!!!!
FigueData
```

```python
#Have attributes in the table

FigueAttr = pd.read_excel(LinkToData,sheet_name='Attributes')
```

```python
import networkx as nx
```

```python
#undirected network
adjacency_u = pd.read_excel(LinkToData,
                            index_col=0, # VERY IMPORTANT!!!!!!!!!!!!!!!!!
                            sheet_name='Relationships')
# see the data frame
adjacency_u
```

```python
graph_adjacency_u = nx.from_pandas_adjacency(adjacency_u)
graph_adjacency_u
```

```python
#undirected network (symmetric relationship)
adjacency_u = pd.read_excel(LinkToData,
                            index_col=0, # VERY IMPORTANT!!!!!!!!!!!!!!!!!
                            sheet_name='Relationships')
# see the data frame
adjacency_u
```

```python
#Building dictionary for attributes
{n:s for n,s in zip(FigueAttr.node,FigueAttr.Multinacional)}
```

```python
FigueAttr_Multinacional={n:s for n,s in zip(FigueAttr.node,FigueAttr.Multinacional)}
nx.set_node_attributes(graph_adjacency_u, FigueAttr_Multinacional, "Multinacional")
```

```python
graph_adjacency_u.nodes.data()
G = graph_adjacency_u
```

```python
import networkx as nx
#Find missing folder path. 
import os

examplesFolder = "Graph/"
os.makedirs(examplesFolder, exist_ok=True)  # Create the folder if it doesn't exist

nx.write_graphml(G, examplesFolder + "css_G.graphml", named_key_ids=True)
```
